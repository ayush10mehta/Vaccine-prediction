# Let's see how the features are correlated with the target variables. 

```python
# This function will help us in creating the bar graph showing the relationship between the features and variables

def vaccination_rate_plot(col, target, data, ax=None):
#col- The column name "feature name"
#target- the 2 targets
#data- the joined data frame

    counts = (joined_df[[target, col]]              #will create a count df which will create table between target and col 
                  .groupby([target, col])
                  .size()
                  .unstack(target)
             )
    group_counts = counts.sum(axis='columns')       #will sum up all the columns value
    props = counts.div(group_counts, axis='index')  #dividing each count will normalize the value thus will give the actual representation

    props.plot(kind="barh", stacked=True, ax=ax)
    ax.invert_yaxis()
    ax.legend().remove()
```

```python
cols_to_plot = [                        #List of cols
    'h1n1_concern',
    'h1n1_knowledge',
    'opinion_h1n1_vacc_effective',
    'opinion_h1n1_risk',
    'opinion_h1n1_sick_from_vacc',
    'opinion_seas_vacc_effective',
    'opinion_seas_risk',
    'opinion_seas_sick_from_vacc',
    'sex',
    'age_group',
    'race',
]

fig, ax = plt.subplots(
    len(cols_to_plot), 2, figsize=(10,len(cols_to_plot)*3.5)
)
for idx, col in enumerate(cols_to_plot):                   #loop for calling the function
    vaccination_rate_plot(
        col, 'h1n1_vaccine', joined_df, ax=ax[idx, 0]
    )
    vaccination_rate_plot(
        col, 'seasonal_vaccine', joined_df, ax=ax[idx, 1]
    )
    
ax[0, 0].legend(
    loc='lower center', bbox_to_anchor=(0.5, 1.05), title='h1n1_vaccine'
)
ax[0, 1].legend(
    loc='lower center', bbox_to_anchor=(0.5, 1.05), title='seasonal_vaccine'
)
fig.tight_layout()
```

# Importing model libraries

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LogisticRegression
from sklearn.multioutput import MultiOutputClassifier

from sklearn.pipeline import Pipeline

from sklearn.model_selection import train_test_split

from sklearn.metrics import roc_curve, roc_auc_score

RANDOM_SEED = 6    # Set a random seed for reproducibility!
```

# Checking the object type of the features

```python
df_Features.dtypes != "object"
```

# Considering only the numeric one

```python
numeric_cols = df_Features.columns[df_Features.dtypes != "object"].values
print(numeric_cols)
```

# Creating pipline

## 1

```python

numeric_preprocessing_steps = Pipeline([
    ('standard_scaler', StandardScaler()),
    ('simple_imputer', SimpleImputer(strategy='median'))
])


preprocessor = ColumnTransformer(
    transformers = [
        ("numeric", numeric_preprocessing_steps, numeric_cols)
    ],
    remainder = "drop"
)
```

## 2

```python
estimators = MultiOutputClassifier(
    estimator=LogisticRegression(penalty="l2", C=1)
)
```

# Final pipeline

```python
full_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("estimators", estimators),
])
```

```python
full_pipeline
```

# Splitting the data 

```python
X_train, X_test, y_train, y_test = train_test_split(
    df_Features,
    df_Labels,
    test_size=0.33,
    shuffle=True,
    stratify=df_Labels,
    random_state=RANDOM_SEED
)
```

# Fitting the model

```python
%%time

# Train model
full_pipeline.fit(X_train, y_train)

# Predict on evaluation set
# This competition wants probabilities, not labels
preds = full_pipeline.predict_proba(X_test)
preds
```

```python
print("test_probas[0].shape", preds[0].shape)
print("test_probas[1].shape", preds[1].shape)
```

# Creating a dataframe of predicted data

```python
y_preds = pd.DataFrame(
    {
        "h1n1_vaccine": preds[0][:, 1],
        "seasonal_vaccine": preds[1][:, 1],
    },
    index = y_test.index
)
print("y_preds.shape:", y_preds.shape)
y_preds.head()
```

# Ploting the ROC curve 

```python
def plot_roc(y_test, y_preds, label_name, ax):
    fpr, tpr, thresholds = roc_curve(y_test, y_preds)
    ax.plot(fpr, tpr)
    ax.plot([0, 1], [0, 1], color='grey', linestyle='--')
    ax.set_ylabel('TPR')
    ax.set_xlabel('FPR')
    ax.set_title(
        f"{label_name}: AUC = {roc_auc_score(y_test, y_preds):.4f}"
    )
```

```python
fig, ax = plt.subplots(1, 2, figsize=(7, 3.5))

plot_roc(
    y_test['h1n1_vaccine'], 
    y_preds['h1n1_vaccine'], 
    'h1n1_vaccine',
    ax=ax[0]
)
plot_roc(
    y_test['seasonal_vaccine'], 
    y_preds['seasonal_vaccine'], 
    'seasonal_vaccine',
    ax=ax[1]
)
fig.tight_layout()
```
